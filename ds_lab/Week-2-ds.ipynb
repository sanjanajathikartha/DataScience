```python
import pandas as pd
import numpy as np

# Sample dataset
df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

print("Original Data:")
print(df)

# Mean Imputation for numerical column
df_mean = df.copy()
df_mean['Age'] = df_mean['Age'].fillna(df_mean['Age'].mean())

# Mode Imputation for categorical column
df_mean['Department'] = df_mean['Department'].fillna(df_mean['Department'].mode()[0])

print("\nAfter Mean/Mode Imputation:")
print(df_mean)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

print("Original Data:")
print(df)

# Drop rows containing missing values
df_drop_rows = df.dropna()

print("\nAfter Dropping Rows:")
print(df_drop_rows)

# Drop columns containing missing values
df_drop_cols = df.dropna(axis=1)

print("\nAfter Dropping Columns:")
print(df_drop_cols)
```

```python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.DataFrame({
    'Salary': [25000, 27000, 29000, 31000, 33000, 35000, 200000]
})

print("Original Data:")
print(df)

z_scores = np.abs(stats.zscore(df['Salary']))

print("\nZ-Scores:")
print(z_scores)

outliers = df[z_scores > 3]

print("\nDetected Outliers:")
print(outliers)

# Remove outliers
df_no_outliers = df[z_scores <= 3]

print("\nData After Removing Outliers:")
print(df_no_outliers)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Salary': [25000, 27000, 29000, 31000, 33000, 35000, 200000]
})

Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Salary'] < lower_bound) | (df['Salary'] > upper_bound)]

print("Detected Outliers:")
print(outliers)

df_clean = df[(df['Salary'] >= lower_bound) & (df['Salary'] <= upper_bound)]

print("\nData After Removing Outliers:")
print(df_clean)
```
